```python
sale_trends_df = pd.read_csv('sale_trends_data_frame.csv', index_col = 0)
```

```python
sale_trends_df = sale_trends_df.fillna(0)
```

```python
sale_trends_df['avg_homeSaleCount'] = ((sale_trends_df['homeSaleCount_2020']
                                        + sale_trends_df['homeSaleCount_2021']
                                        + sale_trends_df['homeSaleCount_2022']
                                        + sale_trends_df['homeSaleCount_2023']
                                        + sale_trends_df['homeSaleCount_2024']
                                        + sale_trends_df['homeSaleCount_2025'])/5).round(2)
sale_trends_df['avg_medSalePrice'] = ((sale_trends_df['medSalePrice_2020']
                                        + sale_trends_df['medSalePrice_2021']
                                        + sale_trends_df['medSalePrice_2022']
                                        + sale_trends_df['medSalePrice_2023']
                                        + sale_trends_df['medSalePrice_2024']
                                        + sale_trends_df['medSalePrice_2025'])/5).round(2)
```

```python
sale_trends_df['avg_homeSaleCount'].quantile(0.75)
```

```python
sale_trends_df.sort_values('avgSalePrice_2025')
```

```python
sale_trends_df.info()
```

```python
df_clean = sale_trends_df[(sale_trends_df['homeSaleCount_2022'] >= 10) & (sale_trends_df['homeSaleCount_2025'] >= 10)].copy()
```

```python
df_clean['cagr_med_price'] = (((df_clean['medSalePrice_2025'] / df_clean['medSalePrice_2022']) ** (1/3) - 1) * 100).round(4)
df_clean['cagr_sale_count'] = (((df_clean['homeSaleCount_2025'] / df_clean['homeSaleCount_2022']) ** (1/3) - 1) * 100).round(4)
```

```python
df_clean = df_clean[df_clean['cagr_med_price']>4.4].sort_values('cagr_med_price', ascending = False)
```

```python
q90 = df_clean['avg_medSalePrice'].quantile(0.9)
q10 = df_clean['avg_medSalePrice'].quantile(0.1)
df_clean = df_clean[(df_clean['avg_medSalePrice'] > q10) & (df_clean['avg_medSalePrice'] < q90)]
```

```python
sale_trends_df_clean = df_clean[['geo_id','medSalePrice_2025', 'homeSaleCount_2025', 'cagr_med_price', 'cagr_sale_count']]
```
