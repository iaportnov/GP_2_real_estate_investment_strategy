```python
def get_all_states():

  url = 'https://api.gateway.attomdata.com/v4/area/state/lookup'

  response = requests.get(url, headers=headers)

  data = response.json()
  states = {x['name']: x['geo_id'] for x in data['response']['result']['package']['item']}
  del states['Guam'], states['American Samoa'],\
   states['Commonwealth of the Northern Mariana Islands'], \
   states['Puerto Rico'], states['United States Virgin Islands']

  return states
```

```python
def get_all_neighborhoods(state_id):
  print(state_id + ' Начало')
  url = f'https://api.gateway.attomdata.com/v4/area/geoid/lookup/'
  neighborhoods = {}
  page = 0
  while True:
    params = {
        'geoId': state_id,
        'GeoType': 'N2',
        'page': page,
        }
    response = requests.get(url, headers=headers, params = params)
    data = response.json()
    neighborhoods_items = {x['geoIdV4']: [x['name'], x['geo_center_latitude'], x['geo_center_longitude'], x['area']] for x in data['response']['result']['package']['item']}
    neighborhoods.update(neighborhoods_items)
    if len(neighborhoods_items) < 200:
      break
    page+=1
  print(state_id + ' Конец')
  return neighborhoods
```
